## 데이터수집

````python
from sklearn.datasets  import load_iris # datasets은 샘플데이터 모음
load_iris = load_iris() # 가져온 샘플데이터를 변수에 저장 
iris_data = load_iris.data # 샘플데이터에는 설명, 데이터, 정답이 다 들어있는데 데이터만 추출
iris_data  #X값 - 데이터이름
````

````python
load_iris.target_names # 예측하려는 '정답 이름" (품종이름)
load_iris.feature_names # 예측하려는 "특성이름"(꽃의 측정값 이름)
````

✅ 뜻:

✅ 결과:

X_train: 학습할 문제(80%)

y_train: 학습할 정답

✅ 비유
test_size=0.2 → 케이크의 20%를 맛보기용으로 잘라두기
random_state=42 → 같은 칼질 위치 기억하기

# 교차검증

교차검증 비유
시험을 여러 세트로 나눠서 "여러 번 시험 보기" 준비   
→ 이걸 왜 하냐면 우연에 의존하지 않고 진짜 실력을 정확하게 평가할 수 있음


### KFold 교차검증

✅ K-Fold 교차검증:   
데이터를 K조각으로 나눠서 → K번 학습/검증을 반복

✅ 예 (K=5):   
5등분 → 각 조각이 한 번씩 “시험 세트”가 됨   
나머지 4조각은 “공부 세트”

✅ 목적:   
모델이 데이터에 “운빨 없이” 잘 맞추는지 평가

````python
from sklearn.model_selection import KFold # 교차검증을 위한 도구 KFold를 호출
kfold = KFold(n_splits=5) # 몇번 나눠서 시험볼 것인지 
iris_clf = DecisionTreeClassifier(random_state=156) # 결정나무분류기 생성 (숫자는 그냥 아무거나 해도 상관없음 )
iris_data, iris_label
````

````python
for train_index, test_index in kfold.split(iris_data):
    X_train, X_test = iris_data[train_index], iris_data[test_index]
    y_train, y_test = iris_label[train_index], iris_label[test_index]
    
    iris_clf.fit(X_train, y_train)
    pred = iris_clf.predict(X_test)
    accuracy = accuracy_score(y_test, pred)
    print(f"Accuracy: {accuracy:.4f}") # 설명은 아래
````

``` for train_index, test_index in kfold.split(iris_data): ```

✅ 뜻

kfold.split() → 데이터를 5개로 나누는 역할   
kfold는 KFold(n_splits=5)로 만들었지?   
split()은 몇 번째 샘플을 학습에 쓰고, 테스트에 쓸지 인덱스를 알려줘

✅ 예시

첫 번째 반복:
train_index = [행번호 1,2,3,4,...]   
test_index = [행번호 5,6,...]

<br>
<br>

``` X_train, X_test = iris_data[train_index], iris_data[test_index] ```   
``` y_train, y_test = iris_label[train_index], iris_label[test_index] ```

✅ 뜻

iris_data(문제들)에서 → 학습용, 테스트용 나누기   
train_index가 가리키는 행만 선택 → X_train   
test_index가 가리키는 행만 선택 → X_test

✅ 비유   
문제집에서 “이번에 공부할 문제” / “시험 볼 문제” 나누기



# stratified split


<details>
<summary>Stratified K-Fold 의 필요성</summary>

K-Fold는 데이터를 k조각으로 나눠서 k번 학습/검증을 반복

**하지만** 문제가 있어!

K-Fold는 그냥 랜덤하게 나눠버려서...
클래스 비율이 불균형해질 수 있어!

***예를 들어, 붓꽃 품종 데이터셋이 이렇게 생겼다고 해보자:***   

| 클래스(품종)    | 샘플 수 |
| ---------- | ---- |
| setosa     | 50   |
| versicolor | 50   |
| virginica  | 50   |


전체 데이터 → 150개 샘플   
클래스별 균형이 잘 맞아있어 (50:50:50)   

여기서 그냥 K-Fold로 나누면?
랜덤하게 나눌 때 → 어떤 Fold에는 setosa가 거의 없거나 너무 많을 수도 있어!

✅ 결과
학습 데이터나 테스트 데이터에서 클래스 비율이 깨짐   
모델이 편향된 학습 → 공정한 평가 어려움

✅ 아주 쉽게:   
시험 문제에 "수학 문제만 가득" → 국어는 연습 안 함

</details>

Stratified K-Fold 의 필요성

````python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(iris_df, iris_df['label']):
  y_train = iris_df['label'].iloc[train_index]
  y_test =  iris_df['label'].iloc[test_index]
  print(f'train 분포: {y_train.value_counts()}')
  print(f'test 분포: {y_test.value_counts()}')
````

````python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)

for train_index, test_index in skf.split(iris_df, iris_df['label']):
  train_data = iris_df.iloc[train_index]
  test_data = iris_df.iloc[test_index]
  
  X_train = train_data[iris.feature_names]
  X_test = test_data[iris.feature_names]
  
  y_train = train_data['label']
  y_test =  test_data['label']
  
  model = DecisionTreeClassifier(random_state=156)
  model.fit(X_train, y_train)
  pred = model.predict(X_test)
  acc = accuracy_score(y_test, pred)
  print(f'acc : {acc}')  
````
