# 새로운 데이터 셋


```python
import seaborn as sns
tips_df = sns.load_dataset('tips')
tips_df.head()
```


y축은 사람 수 
x축은 달러 ( 지불금액, 팁 )

해석결과
어디에 많이 몰려있는 가를 볼 수 있지만 상관관계를 확인할 수가 없다 
ex) 많이 낸 사람이 많은 팁을 준다 

```python
# 일변량 그래프 : 히스토그램 

plt.hist(data=tips_df, x = 'total_bill', bins=20)
```

```python
fig = plt.figure(figsize=(12,5)) # 도화지 크기 설정
axes1 = fig.add_subplot(1,2,1) # 도화지 구성 나누기 1행에 2개의 데이터중 첫번째
axes2 = fig.add_subplot(1,2,2) # 도화지 구성 나누기 1행에 2개의 데이터중 두번째
axes1.hist(data=tips_df, x='total_bill', bins=20)
axes2.hist(data=tips_df, x = 'tip', bins=20)
plt.tight_layout()
plt.show()
```

```python
#boxplot

fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,2,1) #total bill
axes.boxplot(data=tips_df, x ='total_bill')
plt.show() 
```

```python
fig = plt.figure(figsize=(12,5)) # 도화지 확보 
axes = fig.add_subplot(1,1,1) #도화지에다가 나눌 건데(add_subplot) 1,1,1()
axes.boxplot(
    x = [
        tips_df[tips_df['sex'] == 'Female']['total_bill'],
        tips_df[tips_df['sex'] == 'Male']['total_bill']
    ],
    labels=['Female', 'Male']
)
plt.show()
```

female이 지불하는 금액이 일정하다
male은 지불하는 금액 폭이 넓다

최솟값 여성이 더 낮고 남성은 더 높다 -> 여성은 혼자오는데 남성은 둘이 오는 거 같다

```python
fig = plt.figure(figsize=(12,5)) # 도화지 확보 
axes1 = fig.add_subplot(1,2,1) #도화지에다가 나눌 건데(add_subplot) 1,1,1()
axes2 = fig.add_subplot(1,2,2)
axes1.boxplot(
    x = [
        tips_df[(tips_df['sex'] == 'Female') & (tips_df['size'] == 2)]['tip'],
        tips_df[(tips_df['sex'] == 'Male')& (tips_df['size'] == 2)]['tip']
    ],
    labels=['Female', 'Male']
)
axes2.boxplot(
    x = [
        tips_df[(tips_df['sex'] == 'Female') & (tips_df['size'] == 3)]['tip'],
        tips_df[(tips_df['sex'] == 'Male')& (tips_df['size'] == 3)]['tip']
    ],
    labels=['Female', 'Male']
)
plt.show()
```

```python
# scatter, 산점도
```

```python
fig = plt.figure(figsize=(12,12)) # 도화지 확보 
axes = fig.add_subplot(1,1,1) #도화지에다가 나눌 건데(add_subplot) 1,1,1()
axes.scatter(
    tips_df['total_bill'], tips_df['tip']
)
plt.show()
```

```python
colors = {"Female" : "red", "Male":"blue"}
# 다변량 표현
# 성별에 따라 - 색상
# 값의 크기에 따라 점의 크기 

fig = plt.figure() # 도화지 확보 
axes = fig.add_subplot(1,1,1) #도화지에다가 나눌 건데(add_subplot) 1,1,1()
axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s = tips_df['size']**2*10,
    c = tips_df['sex'].map(colors)
)
plt.show()
```

```python
colors = {"Female" : "red", "Male":"blue"}
tips_df['tip_pct'] = tips_df['tip']/tips_df['total_bill']
# 다변량 표현
# 성별에 따라 - 색상
# 값의 크기에 따라 점의 크기 

fig = plt.figure() # 도화지 확보 
axes = fig.add_subplot(1,1,1) #도화지에다가 나눌 건데(add_subplot) 1,1,1()
axes.scatter(
    tips_df['total_bill'], tips_df['size'],
    s = tips_df['tip_pct']**-10*100,
    c = tips_df['sex'].map(colors)
)
plt.show()
```
